```python
import pandas as pd
```

```python
input_path = "../../data/test/sampling_check_by_gemini.csv"
output_path = "../../data/test/sampling_check_by_gemini.csv"
```

```python
# CSVファイルを読み込み、構造を確認する関数
def check_csv_structure(file_path, output_file):
    try:
        # ファイルを読み込み
        df = pd.read_csv(file_path, encoding="utf-8")
        
        # "rule32" カラムを削除
        if "rule32" in df.columns:
            df = df.drop(columns=["rule32"])
            print("rule32カラムを削除しました。")
        
        # データフレームの情報を表示
        print("\n[CSVファイルの構造]")
        df.info()
        
        # 最初の数行を表示
        print("\n[CSVファイルの先頭データ]")
        print(df.head())
        
        # 修正後のCSVファイルを保存
        df.to_csv(output_file, index=False, encoding="utf-8")
        print(f"修正後のCSVファイルを保存しました: {output_file}")
        
    except Exception as e:
        # エラーが発生した場合にエラーメッセージを表示
        print(f"エラー: {e}")

# 関数を実行して確認
check_csv_structure(input_path, output_path)
```
